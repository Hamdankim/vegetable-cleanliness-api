```python
import os
import cv2
import joblib
from src.utils import preprocess, extract_glcm_features

X = []
y = []

PROC = "data/processed/"

for file in os.listdir(PROC):
    if "_blur" not in file:
        continue
    
    # Labeling otomatis berdasarkan nama file
    if "bersih" in file:
        label = 0
    else:
        label = 1

    img = cv2.imread(PROC + file)

    _, gray, _ = preprocess(img)
    features = extract_glcm_features(gray)

    X.append(features)
    y.append(label)

from sklearn.svm import SVC

model = SVC(kernel="linear", C=1.0)
model.fit(X, y)

joblib.dump(model, "models/svm_model.pkl")
print("Model berhasil disimpan di models/svm_model.pkl")
```
