```python
import cv2
import os
import numpy as np
from math import log10, sqrt
from src.utils import preprocess

RAW = "data/raw/"
PROC = "data/processed/"

def mse(img1, img2):
    return np.mean((img1 - img2) ** 2)

def psnr(img1, img2):
    mse_val = mse(img1, img2)
    if mse_val == 0:
        return 100
    return 20 * log10(255.0 / sqrt(mse_val))

for file in os.listdir(RAW):
    path = RAW + file

    img = cv2.imread(path)
    resized, gray, blur = preprocess(img)

    cv2.imwrite(PROC + file.replace(".jpg", "_blur.jpg"), blur)

    print(f"{file} | PSNR: {psnr(gray, blur)} | MSE: {mse(gray, blur)}")
```
